```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Copa America

df_copaamerica = pd.read_csv('/content/Copa_2024_Matches.csv')

# Let's compare the min, max, average, range, variance, standard varation

teams_goals = {

}




df_copaamerica.head()
```

```python
# Adding the goals of each team for away amd home team
for x in df_copaamerica["home_team"].unique():
    teams_goals[x] = df_copaamerica["home_goals"].loc[df_copaamerica["home_team"] == x].sum()
    teams_goals[x] += (df_copaamerica["away_goals"].loc[df_copaamerica["away_team"] == x].sum())



print(teams_goals)
```

```python
# Let's take all of the offensive stats to calculate who was the best offensive
# team in the tournement


"""The formula we will use is the following:
Home_shots + away_shots = s
home_chances + away_changes = c
passes_home + passes_away = p

home_long_pass + away_long_pass = lp
cross_home + away_cross = cp
home_throw + away_throw = tp

home_duels + away duals = d


home_offsides + away_offsides = o
home_yellow + away_yellow = y
home_red + away_red = r

formula_offensive  = ((s + c + p + lp + cp + tp + d + sd - (o + r + y ) ) / 8 ) * 100

  """
teams_offensive_stats = {

}
```

```python
# Total Shots by teams
total_shots ={

}
for x in df_copaamerica["home_team"].unique():
    total_shots[x] = df_copaamerica["Home Total shots"].loc[df_copaamerica["home_team"] == x].sum()
    total_shots[x] += (df_copaamerica["Away Total shots"].loc[df_copaamerica["away_team"] == x].sum())

print(f"Shots by team ==> {total_shots}")

# Total chances by teams

total_chances ={

}

for x in df_copaamerica["home_team"].unique():
    total_chances[x] = df_copaamerica["Home Big chances"].loc[df_copaamerica["home_team"] == x].sum()
    total_chances[x] += (df_copaamerica["Away Big chances"].loc[df_copaamerica["away_team"] == x].sum())

print(f"Chances by team ==> {total_chances}")

# Total passes by teams

total_passes ={

}

for x in df_copaamerica["home_team"].unique():
    total_passes[x] = df_copaamerica["Home Passes"].loc[df_copaamerica["home_team"] == x].sum()
    total_passes[x] += (df_copaamerica["Away Passes"].loc[df_copaamerica["away_team"] == x].sum())

print(f"Passes by teams ===> {total_passes}")
```

```python
Corners = {

}
for x in df_copaamerica["home_team"].unique():
    Corners[x] = df_copaamerica["Home Corners"].loc[df_copaamerica["home_team"] == x].sum()
    Corners[x] += (df_copaamerica["Away Corners"].loc[df_copaamerica["away_team"] == x].sum())

print(f"Long Passes & Corners by teams ==> {Corners}")




throws = {

}

for x in df_copaamerica["home_team"].unique():
    throws[x] = df_copaamerica["Home Throws"].loc[df_copaamerica["home_team"] == x].sum()
    throws[x] += (df_copaamerica["Away Throws"].loc[df_copaamerica["away_team"] == x].sum())


print(f"Throws in by teams ==> {throws}")


duels_won = {

}

for x in df_copaamerica["home_team"].unique():
    duels_won[x] = df_copaamerica["Home Duels won"].loc[df_copaamerica["home_team"] == x].sum()
    duels_won[x] += (df_copaamerica["Away Duels won"].loc[df_copaamerica["away_team"] == x].sum())

print(f"Duels won by teams ==> {duels_won}")



```

```python
# Fouls + Red + Yellow

red_yellows_fouls = {

}

for x in df_copaamerica["home_team"].unique():
    red_yellows_fouls[x] = df_copaamerica["Home Fouls committed"].loc[df_copaamerica["home_team"] == x].sum()
    red_yellows_fouls[x] += (df_copaamerica["Away Fouls committed"].loc[df_copaamerica["away_team"] == x].sum())
    red_yellows_fouls[x] += (df_copaamerica["Home Red cards"].loc[df_copaamerica["home_team"] == x].sum())
    red_yellows_fouls[x] += (df_copaamerica["Away Red cards"].loc[df_copaamerica["away_team"] == x].sum())
    red_yellows_fouls[x] += (df_copaamerica["Home Yellow cards"].loc[df_copaamerica["home_team"] == x].sum())
    red_yellows_fouls[x] += (df_copaamerica["Away Yellow cards"].loc[df_copaamerica["away_team"] == x].sum())

print(f"Red + Yellow + Fouls by teams ==> {red_yellows_fouls}")






offsides = {

}

for x in df_copaamerica["home_team"].unique():
    offsides[x] = df_copaamerica["Home Offsides"].loc[df_copaamerica["home_team"] == x].sum()
    offsides[x] += (df_copaamerica["Away Offsides"].loc[df_copaamerica["away_team"] == x].sum())

print(f"Offsides by teams ==> {offsides}")
```

```python
# Let's calculate the offensive stats by each team

for x in df_copaamerica["home_team"].unique():
  teams_offensive_stats[x] = 0

  teams_offensive_stats[x] += (total_shots[x] + total_chances[x] + total_passes[x] + Corners[x] + throws[x] + duels_won[x] - (red_yellows_fouls[x] + offsides[x]))
  teams_offensive_stats[x] /= 16



print(f"Offensive stats ==> {teams_offensive_stats}")
```

```python
"""
Let's calculate the deffesinve stats for each team

home_fouls + away_fouls = f
home_interceptions + away_interceptions = i
home_blocks + away_blocks = b
home_clearances + away_clearances = c
home_keeper + away_keeper = k
home_woodwork + away_woodwork = w

formula_defensive = ((f  + i + b + c + k + w) / 7) * 100
"""

fouls_team = {

}

for x in df_copaamerica["home_team"].unique():
    fouls_team[x] = df_copaamerica["Home Fouls committed"].loc[df_copaamerica["home_team"] == x].sum()
    fouls_team[x] += (df_copaamerica["Away Fouls committed"].loc[df_copaamerica["away_team"] == x].sum())

print(f"Fouls by teams ==> {fouls_team}")


interceptions_team = {

}

for x in df_copaamerica["home_team"].unique():
    interceptions_team[x] = df_copaamerica["Home Interceptions"].loc[df_copaamerica["home_team"] == x].sum()
    interceptions_team[x] += (df_copaamerica["Away Interceptions"].loc[df_copaamerica["away_team"] == x].sum())


print(f"Interceptions by teams ==> {interceptions_team}")



blocks_team = {

}


for x in df_copaamerica["home_team"].unique():
    blocks_team[x] = df_copaamerica["Home Blocks"].loc[df_copaamerica["home_team"] == x].sum()
    blocks_team[x] += (df_copaamerica["Away Blocks"].loc[df_copaamerica["away_team"] == x].sum())


print(f"Blocks by teams ==> {blocks_team}")
```

```python
cleareances_team = {  }

teams_defensive_stats = {

}

for x in df_copaamerica["home_team"].unique():
    cleareances_team[x] = df_copaamerica["Home Clearances"].loc[df_copaamerica["home_team"] == x].sum()
    cleareances_team[x] += (df_copaamerica["Away Clearances"].loc[df_copaamerica["away_team"] == x].sum())


print(f"Cleareances by teams ==> {cleareances_team}")

keeps = {}


for x in df_copaamerica["home_team"].unique():
    keeps[x] = df_copaamerica["Home Keeper saves"].loc[df_copaamerica["home_team"] == x].sum()
    keeps[x] += (df_copaamerica["Away Keeper saves"].loc[df_copaamerica["away_team"] == x].sum())


print(f"Keeper throws by teams ==> {keeps}")


woodwork = {}


for x in df_copaamerica["home_team"].unique():
    woodwork[x] = df_copaamerica["Home Hit woodwork"].loc[df_copaamerica["home_team"] == x].sum()
    woodwork[x] += (df_copaamerica["Away Hit woodwork"].loc[df_copaamerica["away_team"] == x].sum())


print(f"Woodwork by teams ==> {woodwork}")

# Adding all of the defensive stats for each team

for x in df_copaamerica["home_team"].unique():
  teams_defensive_stats[x] = 0
  teams_defensive_stats[x] += (fouls_team[x] + interceptions_team[x] + blocks_team[x] + cleareances_team[x] + keeps[x] + woodwork[x])
  teams_defensive_stats[x] /= 11



print(f"Defensive stats ==> {teams_defensive_stats}")



```

```python
# Creating a visualizing for the best teams of the copa america



plt.style.use("_mpl-gallery")
np.random.seed(0)

x = list(teams_offensive_stats.keys())
y = list(teams_offensive_stats.values())

print(teams_offensive_stats)

sizes = np.random.uniform(20, 80, len(x))
colors = np.random.uniform(50, 90, len(x))

fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)




ax.set(xlim=(4, len(teams_offensive_stats.keys())), xticks=np.arange(-2, len(teams_offensive_stats.keys())+1),
       ylim=(0, 300), yticks=np.arange(0, 300, 15))

fig.set_size_inches(18, 8, forward=True)


plt.scatter(x, y, s=sizes, c=colors, vmin=1, vmax=100)
plt.xticks(rotation=10)

plt.plot(x, y, '.r-')
plt.show()
```
